```python
# --- 1. Import the handler ---
from reaxkit.io.ffield_handler import FFieldHandler

# --- 2. Load the ffield file ---
handler = FFieldHandler("ffield")  # adjust path if needed

# --- 3. Force parsing ---
df_summary = handler.dataframe()   # should be empty
print("Summary DF:")
display(df_summary)

# --- 4. Show available sections ---
print("Sections detected:", list(handler.sections.keys()))

# --- 5. Display each section one-by-one ---

print("\n=== GENERAL PARAMETERS ===")
display(handler.general_df)

print("\n=== ATOM PARAMETERS ===")
display(handler.atom_df)

print("\n=== BOND PARAMETERS ===")
display(handler.bond_df)

print("\n=== OFF-DIAGONAL PARAMETERS ===")
display(handler.off_diagonal_df)

print("\n=== ANGLE PARAMETERS ===")
display(handler.angle_df)

print("\n=== TORSION PARAMETERS ===")
display(handler.torsion_df)

print("\n=== HBOND PARAMETERS ===")
display(handler.hbond_df)
```

```python
print("Atom columns:", handler.atom_df.columns)
print("Bond columns:", handler.bond_df.columns)
print("Torsion columns:", handler.torsion_df.columns)
```

```python
#using ffield analyzer

from reaxkit.io.ffield_handler import FFieldHandler
from reaxkit.analysis.ffield_analyzer import get_sections_data

handler = FFieldHandler("ffield")

print(handler.metadata().get("description"))

general_df = get_sections_data(handler, section="general")
atom_df    = get_sections_data(handler, section="atoms")
bond_df    = get_sections_data(handler, section="bonds")

display(general_df.head())
display(atom_df.head())
display(bond_df.head())
```
