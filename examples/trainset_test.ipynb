```python
from reaxkit.io.trainset_handler import TrainsetHandler
from reaxkit.analysis.trainset_analyzer import trainset_group_comments

handler = TrainsetHandler("trainset.in")

# show what sections exist
print("Sections:", handler.metadata()["sections"])

# get CHARGE section table
try:
    charges = handler.charges()
    print(charges.head())
except:
    print('there are no charges')

# get ENERGY section
energy = handler.energy_terms()
print(energy)
energy.to_csv("trainset_energy.csv")

# get any section by name
try:
    geom = handler.section("GEOMETRY")
    print(geom)
except:
    print("There are no geometry sections")
```
